Finally, we will assume that line losses and trade costs are zero such that the marginal willingness to pay and marginal generation costs of "transmission lines" do not have to be corrected. 

To run the model, we need to specify the marginal generation costs of electrolyzers, which are described more thoroughly in the home exercise Project-H. Electrolyzer are included in the set $\mathcal{I}^{HP}$ and are therefore modelled identically to a heat pump. The electrolyzer plant produces hydrogen $(H_2)$ by splitting the hydrogen elements from water $(H_2O)$. It does so using electricity $(E)$ as an energy input and the only production residues are oxygen $(O)$ and heat $(H)$, whith the latter being distributed into the district heating system. 

Given the setup in Project-H, it is possible to derive, that electrolyzers' marginal willinges to pay for electricity is given by
\begin{align}
    \mu^E \equiv \frac{1}{\nu_H}\left(c-\lambda^H-\lambda^{H_2}\frac{\nu_H}{\nu_{H_2}}\right).
\end{align}, where
- $\nu_H<0$ is an electricity-to-heat ratio, describing the electricity input per heat output, i.e. $E_h = \nu_H H_h$. 
- $\nu_{H_2}<0$ is an electricity-to-hydrogen ratio, describing the electricity input per hydrogen output, i.e. $E_h = \nu_{H_2} {H_2}_h$.
- $(c)$ is the constant marginal operation and maintanence costs of heat generation.
- Finally, $\lambda^E$, $\lambda^H$, and $\lambda^{H_2}$ are the (shadow) prices for the electricity, heat, and hydrogen price, respectively.


**Assume that the hydrogen market is represented by a perfectly elastic demand curve and adjust `OtherMC` to account for the marginal hydrogen revenues pr generated unit of $H$. We are assuming20 DKK/kg $H_2$, which is converted into €/MWh below.**

```python
DKK_kg = 29.84 # Levelised cost of hydrogen
kWh_kg = 39
Euro_DKK = 1/7.45
MWh_kWh = 1/1000
GJ_MWh = 0.277778

λ_H2 = DKK_kg*(1/kWh_kg)*Euro_DKK*(1/MWh_kWh)
```

```python
"""
MOC for full political scenario 
"""

def MOC(type):
    # Combine 'mc' and 'OtherMC'
    combined_mc = db['mc'] + db['OtherMC']

    # Group by 'id' and calculate the mean
    average_mc_per_id = combined_mc.groupby('id').mean()
    
    title = {'H' : 'Heat', 'E' : 'Energy'}
        

    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt

    # Settings
    ϵ = np.finfo(float).eps * 1e10
    lower_point = 0
    upper_point = average_mc_per_id.max() * 1.1  # Set upper_point to 10% above the max price

    # Combine 'mc' and 'OtherMC', and compute the average price by 'id'
    combined_mc = db['mc'] + db['OtherMC']
    average_mc_per_id = combined_mc.groupby('id').mean()

    # Use 'GeneratingCap_E' as generation capacity (q)
    generation_capacity = db['GeneratingCap_'+type].groupby('id').mean()

    # Create DataFrame for merit order curve
    df = pd.DataFrame({
        'q': generation_capacity,
        'p': average_mc_per_id
    }).dropna()  # Drop any rows with missing values

    # Sort and calculate cumulative capacity
    df = df.sort_values('p')
    df['q_cs'] = df['q'].cumsum()

    # Add upper points for step plotting
    df_up = df.copy()
    df_up['q_cs'] = df_up['q_cs'] + ϵ
    df_up['p'] = df_up['p'].shift(-1).fillna(upper_point)
    df = pd.concat([df, df_up], axis=0).sort_values(['p', 'q_cs'])

    # Merit order curve plot
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot the merit order curve
    df.set_index('q_cs')['p'].plot(ax=ax, drawstyle="steps-post", label="Merit Order Curve", color="blue")
    ax.set_title(f'Merit Order Curve for {title[type]} Plants')
    ax.set_xlabel('Cumulative Capacity (MW)')
    ax.set_ylabel('Price (dkk/GJ)')
    ax.grid(True)
    ax.legend()

    # Save and show the figure
    fig.tight_layout()
    plt.show()
```

```python
def MOC(type):
    """
    Plots the Merit Order Curve for both `db` and `dbtake1`, overlaying their results.
    
    Parameters:
    - type: 'H' for Heat or 'E' for Energy
    """
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt

    def prepare_merit_order_curve(database, label, color):
        # Convert 'mc' from EUR/GJ to EUR/MWh
        mc_converted = database['mc'] * GJ_MWh
        
        # Combine 'mc' (converted) and 'OtherMC'
        combined_mc = mc_converted + database['OtherMC']
        
        # Group by 'id' and calculate the mean
        average_mc_per_id = combined_mc.groupby('id').mean()
        
        # Use 'GeneratingCap_E' or 'GeneratingCap_H' based on `type`
        generation_capacity = database['GeneratingCap_' + type].groupby('id').mean()
        
        # Create DataFrame for merit order curve
        df = pd.DataFrame({
            'q': generation_capacity,
            'p': average_mc_per_id
        }).dropna()  # Drop rows with missing values
        
        # Sort and calculate cumulative capacity
        df = df.sort_values('p')
        df['q_cs'] = df['q'].cumsum()
        
        # Add upper points for step plotting
        ϵ = np.finfo(float).eps * 1e10
        upper_point = average_mc_per_id.max() * 1.1
        df_up = df.copy()
        df_up['q_cs'] = df_up['q_cs'] + ϵ
        df_up['p'] = df_up['p'].shift(-1).fillna(upper_point)
        df = pd.concat([df, df_up], axis=0).sort_values(['p', 'q_cs'])
        
        return df, label, color

    # Prepare data for `db` and `dbtake1`
    db_curve, db_label, db_color = prepare_merit_order_curve(db, "With Energy Islands", "blue")
    dbtake1_curve, dbtake1_label, dbtake1_color = prepare_merit_order_curve(dbtake1, "Without Energy Islands", "orange")

    # Plot the merit order curves
    fig, ax = plt.subplots(figsize=(10, 6))
    db_curve.set_index('q_cs')['p'].plot(ax=ax, drawstyle="steps-post", label=db_label, color=db_color)
    dbtake1_curve.set_index('q_cs')['p'].plot(ax=ax, drawstyle="steps-post", label=dbtake1_label, color=dbtake1_color)

    # Add titles and labels
    title = {'H': 'Heat', 'E': 'Energy'}
    ax.set_title(f'Merit Order Curve for {title[type]} Plants')
    ax.set_xlabel('Cumulative Capacity (MW)')
    ax.set_ylabel('Price (€/MWh)')  # Updated units in the label
    ax.grid(True)
    ax.legend()

    # Save and show the figure
    fig.tight_layout()
    plt.show()
```

```python
#Reminder of databases thru dict
databases = {
    'S0': dbtake1,
    'S1': db,
    'S2': dbonlybornholm,
    'S3': dbbornholm_full,
    'S4': dball
}

```

```python
# %pip install -q lightgbm
import warnings; warnings.filterwarnings("ignore")

import pandas as pd, numpy as np
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ===== 0) CONFIG =====
START_DATE = pd.Timestamp("2024-01-01 00:00:00")   # <-- set to the real start timestamp
H_TEST_DAYS = 30                                   # hold out last 30 days for quick eval
H_FWD_DAYS  = 548                                  # forecast ~18 months forward
# 'databases' must exist: {'S0': dbtake1, 'S1': db, 'S2': dbonlybornholm, 'S3': dbbornholm_full, 'S4': dball}

# ===== 1) reshape: each scenario -> tidy long dataframe =====
def series_to_long(db, scenario_label):
    """
    Expects db['marginalSystemCosts_E'] as a Series with MultiIndex (zone=g_E, hour=h starting at 1).
    """
    s = db['marginalSystemCosts_E']
    if not isinstance(s, (pd.Series, pd.DataFrame)):
        raise TypeError(f"{scenario_label}: expected Series/DataFrame at key 'marginalSystemCosts_E'")
    if isinstance(s, pd.DataFrame):
        # If someone gave a DataFrame column, collapse it to a Series
        s = s.squeeze('columns')
    s = s.rename('cost')

    if s.index.nlevels != 2:
        raise ValueError(f"{scenario_label}: expected MultiIndex (zone, hour). Got levels={s.index.nlevels}")

    zones = s.index.get_level_values(0)
    hours = s.index.get_level_values(1)
    if not np.issubdtype(pd.Series(hours).dtype, np.number):
        # try to coerce hour to integer
        hours = pd.to_numeric(hours, errors='raise')

    ts = START_DATE + pd.to_timedelta(hours - 1, unit='h')
    out = pd.DataFrame({'timestamp': ts, 'zone': zones, 'cost': s.values})
    out['scenario'] = scenario_label
    return out[['timestamp','scenario','zone','cost']].sort_values(['scenario','zone','timestamp'])

long_list = [series_to_long(db, sc) for sc, db in databases.items()]
long = pd.concat(long_list, ignore_index=True).sort_values(['scenario','zone','timestamp']).reset_index(drop=True)

# ===== 2) feature engineering (safe lags + rolling + calendar/fourier) =====
group_cols = ['scenario','zone']
# find minimal history length across (scenario, zone) to decide valid lags
min_len = long.groupby(group_cols).size().min()

raw_lags = [1, 24, 24*7, 24*30, 24*90, 24*182, 24*365]  # hourly: 1h, 1d, 1w, ~1m, ~3m, ~6m, ~1y
lag_hours = [L for L in raw_lags if L < min_len]        # only use lags that exist

# Lags
for L in lag_hours:
    long[f'lag_{L}'] = long.groupby(group_cols, observed=True)['cost'].shift(L)

# Rolling windows (shift(1) to avoid leakage)
roll_windows = [24, 24*7, 24*30]
for w in roll_windows:
    long[f'roll_mean_{w}'] = (
        long.groupby(group_cols, observed=True)['cost'].shift(1).rolling(w, min_periods=max(8, w//4)).mean()
    )
    long[f'roll_std_{w}'] = (
        long.groupby(group_cols, observed=True)['cost'].shift(1).rolling(w, min_periods=max(8, w//4)).std()
    )

# Calendar / seasonal
dt = long['timestamp']
long['hour']  = dt.dt.hour
long['dow']   = dt.dt.dayofweek
long['month'] = dt.dt.month
# Fourier seasonality
long['sin_day']  = np.sin(2*np.pi*long['hour']/24.0)
long['cos_day']  = np.cos(2*np.pi*long['hour']/24.0)
long['sin_week'] = np.sin(2*np.pi*(long['dow']*24 + long['hour'])/(24*7))
long['cos_week'] = np.cos(2*np.pi*(long['dow']*24 + long['hour'])/(24*7))
long['sin_year'] = np.sin(2*np.pi*dt.dt.dayofyear/365.25)
long['cos_year'] = np.cos(2*np.pi*dt.dt.dayofyear/365.25)

feat = long.dropna().reset_index(drop=True)
feat['scenario'] = feat['scenario'].astype('category')
feat['zone']     = feat['zone'].astype('category')

# ===== 3) train/validation split (last H_TEST_DAYS as holdout) =====
H_TEST = 24 * H_TEST_DAYS
last_ts = feat['timestamp'].max()
cutoff  = last_ts - pd.Timedelta(hours=H_TEST)

feature_cols = [c for c in feat.columns if c not in ['timestamp','cost']]

train = feat[feat['timestamp'] <= cutoff]
test  = feat[feat['timestamp']  > cutoff]

assert train.shape[0] > 0 and len(feature_cols) > 0, f"Empty train/features: train={train.shape}, features={len(feature_cols)}"

# ===== 4) train LightGBM (global across scenarios & zones) =====
model = lgb.LGBMRegressor(
    n_estimators=2000, learning_rate=0.02, num_leaves=256,
    subsample=0.8, colsample_bytree=0.8, reg_lambda=1.0, random_state=42
)
model.fit(train[feature_cols], train['cost'], categorical_feature=['scenario','zone'])

# quick eval
pred_test = model.predict(test[feature_cols])
mae  = mean_absolute_error(test['cost'], pred_test)
rmse = mean_squared_error(test['cost'], pred_test, squared=False)
print(f"Holdout ({H_TEST_DAYS} days)  MAE={mae:.3f}  RMSE={rmse:.3f}")

# ===== 5) iterative multi-step forecast for > 1 year =====
def forecast_iterative(history_df, start_after_ts, hours, model, feature_cols, lag_hours, roll_windows):
    """
    history_df: columns ['timestamp','scenario','zone','cost']
    start_after_ts: last historical timestamp (forecast begins at +1h)
    hours: how many hours forward to predict
    """
    scenarios = history_df['scenario'].unique()
    zones     = history_df['zone'].unique()

    # working state = history (grows with predictions)
    state = history_df.sort_values(['scenario','zone','timestamp']).copy().reset_index(drop=True)
    preds = []

    for t in pd.date_range(start_after_ts + pd.Timedelta(hours=1), periods=hours, freq='H'):
        rows = []
        for sc in scenarios:
            for zn in zones:
                hz = state[(state['scenario']==sc) & (state['zone']==zn)].sort_values('timestamp')

                row = {
                    'timestamp': t, 'scenario': sc, 'zone': zn,
                    'hour': t.hour, 'dow': t.dayofweek, 'month': t.month,
                    'sin_day':  np.sin(2*np.pi*t.hour/24.0),
                    'cos_day':  np.cos(2*np.pi*t.hour/24.0),
                    'sin_week': np.sin(2*np.pi*(t.dayofweek*24 + t.hour)/(24*7)),
                    'cos_week': np.cos(2*np.pi*(t.dayofweek*24 + t.hour)/(24*7)),
                    'sin_year': np.sin(2*np.pi*t.dayofyear/365.25),
                    'cos_year': np.cos(2*np.pi*t.dayofyear/365.25),
                }
                # lags
                vals = hz['cost'].values
                for L in lag_hours:
                    row[f'lag_{L}'] = vals[-L] if len(vals) >= L else np.nan
                # rollings (use last w prior points)
                for w in roll_windows:
                    s = hz['cost'].iloc[-(w+1):-1] if len(hz) > w else hz['cost'].iloc[:-1]
                    row[f'roll_mean_{w}'] = s.tail(w).mean() if len(s) else np.nan
                    row[f'roll_std_{w}']  = s.tail(w).std()  if len(s) else np.nan

                rows.append(row)

        X = pd.DataFrame(rows)
        X['scenario'] = X['scenario'].astype('category')
        X['zone']     = X['zone'].astype('category')
        # basic safety for early steps
        X = X.fillna(method='ffill').fillna(method='bfill')

        yhat = model.predict(X[feature_cols])
        X['pred'] = yhat
        preds.append(X[['timestamp','scenario','zone','pred']])

        # append to state so subsequent steps can use the predictions as lags
        state = pd.concat([state, X[['timestamp','scenario','zone','pred']].rename(columns={'pred':'cost'})],
                          ignore_index=True)

    return pd.concat(preds, ignore_index=True)

# Forecast forward
H_FWD = 24 * H_FWD_DAYS
historical_min = long[['timestamp','scenario','zone','cost']].copy()
future = forecast_iterative(historical_min, last_ts, H_FWD, model, feature_cols, lag_hours, roll_windows)

print("Forecast head:")
print(future.head())

# ===== 6) (optional) quick plot for one scenario+zone =====
try:
    import matplotlib.pyplot as plt
    pick_sc, pick_zn = feat['scenario'].cat.categories[0], feat['zone'].cat.categories[0]
    hist_sel = historical_min[(historical_min['scenario']==pick_sc) & (historical_min['zone']==pick_zn)]
    fut_sel  = future[(future['scenario']==pick_sc) & (future['zone']==pick_zn)]

    plt.figure(figsize=(12,4))
    plt.plot(hist_sel['timestamp'], hist_sel['cost'], label='history')
    plt.plot(fut_sel['timestamp'], fut_sel['pred'], label='forecast')
    plt.title(f"Marginal System Costs — {pick_sc}/{pick_zn}")
    plt.xlabel("Time"); plt.ylabel("€/MWh"); plt.legend(); plt.grid(True); plt.tight_layout()
    plt.show()
except Exception as e:
    print("Plot skipped:", e)
```

```python
# %pip install -q lightgbm
import warnings; warnings.filterwarnings("ignore")

import pandas as pd, numpy as np
import lightgbm as lgb
from lightgbm import early_stopping, log_evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ===== CONFIG =====
START_DATE   = pd.Timestamp("2024-01-01 00:00:00")  # set to your real start
H_VAL_DAYS   = 30                                    # validation window (early stopping)
H_TEST_DAYS  = 30                                    # test window for reporting
H_FWD_DAYS   = 548                                   # forecast horizon (~18 months)
SEASONAL_BLEND = 0.20                                 # 0..1; 0 = no blend, 0.2 = 20% seasonal naive
WEEK_H = 24*7

# ===== reshape all scenarios to one long dataframe =====
def series_to_long(db, scenario_label):
    s = db['marginalSystemCosts_E']
    if isinstance(s, pd.DataFrame): s = s.squeeze('columns')
    s = s.rename('cost')
    zones = s.index.get_level_values(0)
    hours = pd.to_numeric(s.index.get_level_values(1))
    ts = START_DATE + pd.to_timedelta(hours - 1, unit='h')
    out = pd.DataFrame({'timestamp': ts, 'scenario': scenario_label, 'zone': zones, 'cost': s.values})
    return out.sort_values(['scenario','zone','timestamp'])

long = pd.concat([series_to_long(db, sc) for sc, db in databases.items()],
                 ignore_index=True).sort_values(['scenario','zone','timestamp'])

# ensure non-negative for Tweedie
long['cost'] = long['cost'].clip(lower=0)

# ===== features =====
group_cols = ['scenario','zone']
min_len = long.groupby(group_cols).size().min()

raw_lags   = [1, 24, WEEK_H, 24*30, 24*90, 24*182]     # keep lags that exist with 1 year history
lag_hours  = [L for L in raw_lags if L < min_len]
roll_windows = [24, WEEK_H, 24*30]

long = long.sort_values(group_cols + ['timestamp']).copy()
for L in lag_hours:
    long[f'lag_{L}'] = long.groupby(group_cols, observed=True)['cost'].shift(L)
for w in roll_windows:
    long[f'roll_mean_{w}'] = long.groupby(group_cols, observed=True)['cost'].shift(1).rolling(
        w, min_periods=max(8, w//4)
    ).mean()
    long[f'roll_std_{w}']  = long.groupby(group_cols, observed=True)['cost'].shift(1).rolling(
        w, min_periods=max(8, w//4)
    ).std()

dt = long['timestamp']
long['hour']   = dt.dt.hour
long['dow']    = dt.dt.dayofweek
long['month']  = dt.dt.month
long['qtr']    = dt.dt.quarter
long['is_wend']= (long['dow']>=5).astype(int)

# Fourier seasonality
long['sin_hour']  = np.sin(2*np.pi*long['hour']/24.0)
long['cos_hour']  = np.cos(2*np.pi*long['hour']/24.0)
long['sin_week']  = np.sin(2*np.pi*(long['dow']*24 + long['hour'])/(24*7))
long['cos_week']  = np.cos(2*np.pi*(long['dow']*24 + long['hour'])/(24*7))
long['sin_year']  = np.sin(2*np.pi*dt.dt.dayofyear/365.25)
long['cos_year']  = np.cos(2*np.pi*dt.dt.dayofyear/365.25)

feat = long.dropna().reset_index(drop=True)
feat['scenario'] = feat['scenario'].astype('category')
feat['zone']     = feat['zone'].astype('category')

# ===== split: train / val / test (time-based) =====
last_ts  = feat['timestamp'].max()
cut_test = last_ts - pd.Timedelta(hours=24*H_TEST_DAYS)
cut_val  = cut_test - pd.Timedelta(hours=24*H_VAL_DAYS)

feature_cols = [c for c in feat.columns if c not in ['timestamp','cost']]

train = feat[feat['timestamp'] <= cut_val]
val   = feat[(feat['timestamp'] > cut_val) & (feat['timestamp'] <= cut_test)]
test  = feat[feat['timestamp']  > cut_test]

assert train.shape[0] and val.shape[0] and test.shape[0], "Time splits are empty; reduce H_VAL/H_TEST."

# ===== model (Tweedie for spiky, non-negative targets) =====

model = lgb.LGBMRegressor(
    objective='tweedie', tweedie_variance_power=1.3,
    n_estimators=5000, learning_rate=0.02,
    num_leaves=512, min_data_in_leaf=64,
    subsample=0.9, colsample_bytree=0.9,
    reg_lambda=1.0, random_state=42
)

model.fit(
    train[feature_cols], train['cost'],
    categorical_feature=['scenario','zone'],
    eval_set=[(val[feature_cols], val['cost'])],
    eval_metric='l1',  # mae
    callbacks=[
        early_stopping(stopping_rounds=300, first_metric_only=True),
        log_evaluation(period=50)  # set to 0 to silence
    ]
)

# report on test with best_iteration_
pred_val  = model.predict(val[feature_cols],  num_iteration=model.best_iteration_)
pred_test = model.predict(test[feature_cols], num_iteration=model.best_iteration_)
mae_v  = mean_absolute_error(val['cost'],  pred_val)
rmse_v = mean_squared_error(val['cost'],  pred_val, squared=False)
mae_t  = mean_absolute_error(test['cost'], pred_test)
rmse_t = mean_squared_error(test['cost'], pred_test, squared=False)
print(f"Val  ({H_VAL_DAYS}d)  MAE={mae_v:.3f}  RMSE={rmse_v:.3f}")
print(f"Test ({H_TEST_DAYS}d) MAE={mae_t:.3f}  RMSE={rmse_t:.3f}")

# ===== iterative multi-step forecast with optional seasonal blend =====
def forecast_iterative(history_df, start_after_ts, hours, model, feature_cols,
                       lag_hours, roll_windows, seasonal_blend=0.0):
    scenarios = history_df['scenario'].unique()
    zones     = history_df['zone'].unique()
    state = history_df.sort_values(['scenario','zone','timestamp']).copy().reset_index(drop=True)
    out = []

    for t in pd.date_range(start_after_ts + pd.Timedelta(hours=1), periods=hours, freq='H'):
        rows = []
        for sc in scenarios:
            for zn in zones:
                hz = state[(state['scenario']==sc) & (state['zone']==zn)].sort_values('timestamp')

                row = {
                    'timestamp': t, 'scenario': sc, 'zone': zn,
                    'hour': t.hour, 'dow': t.dayofweek, 'month': t.month,
                    'qtr': ((t.month-1)//3)+1, 'is_wend': int(t.weekday()>=5),
                    'sin_hour': np.sin(2*np.pi*t.hour/24.0), 'cos_hour': np.cos(2*np.pi*t.hour/24.0),
                    'sin_week': np.sin(2*np.pi*(t.dayofweek*24 + t.hour)/(24*7)),
                    'cos_week': np.cos(2*np.pi*(t.dayofweek*24 + t.hour)/(24*7)),
                    'sin_year': np.sin(2*np.pi*t.dayofyear/365.25),
                    'cos_year': np.cos(2*np.pi*t.dayofyear/365.25),
                }
                vals = hz['cost'].values
                for L in lag_hours:
                    row[f'lag_{L}'] = vals[-L] if len(vals) >= L else np.nan
                for w in roll_windows:
                    s = hz['cost'].iloc[-(w+1):-1] if len(hz) > w else hz['cost'].iloc[:-1]
                    row[f'roll_mean_{w}'] = s.tail(w).mean() if len(s) else np.nan
                    row[f'roll_std_{w}']  = s.tail(w).std()  if len(s) else np.nan

                # seasonal naive baseline (last week same hour), if available
                row['_sn'] = vals[-WEEK_H] if len(vals) >= WEEK_H else np.nan
                rows.append(row)

        X = pd.DataFrame(rows)
        X['scenario'] = X['scenario'].astype('category')
        X['zone']     = X['zone'].astype('category')
        X = X.fillna(method='ffill').fillna(method='bfill')

        yhat = model.predict(X[feature_cols], num_iteration=model.best_iteration_)
        if seasonal_blend > 0:
            sn = X['_sn'].to_numpy()
            # if _sn is NaN (early few steps), fall back to model-only
            mask = ~np.isnan(sn)
            yhat_blend = yhat.copy()
            yhat_blend[mask] = (1 - seasonal_blend) * yhat[mask] + seasonal_blend * sn[mask]
            yhat = yhat_blend

        X['pred'] = np.maximum(yhat, 0.0)  # keep non-negative
        out.append(X[['timestamp','scenario','zone','pred']])
        state = pd.concat([state, X[['timestamp','scenario','zone','pred']].rename(columns={'pred':'cost'})],
                          ignore_index=True)

    return pd.concat(out, ignore_index=True)

H_FWD = 24 * H_FWD_DAYS
historical_min = long[['timestamp','scenario','zone','cost']].copy()
future = forecast_iterative(
    historical_min, last_ts, H_FWD, model, feature_cols,
    lag_hours, roll_windows, seasonal_blend=SEASONAL_BLEND
)

# quick plot for one series
try:
    import matplotlib.pyplot as plt
    pick_sc = feat['scenario'].cat.categories[0]
    pick_zn = feat['zone'].cat.categories[0]
    hist_sel = historical_min[(historical_min['scenario']==pick_sc) & (historical_min['zone']==pick_zn)]
    fut_sel  = future[(future['scenario']==pick_sc) & (future['zone']==pick_zn)]

    plt.figure(figsize=(14,4))
    plt.plot(hist_sel['timestamp'], hist_sel['cost'], label='history')
    plt.plot(fut_sel['timestamp'], fut_sel['pred'], label='forecast')
    plt.title(f"Marginal System Costs — {pick_sc}/{pick_zn}")
    plt.ylabel("€/MWh"); plt.xlabel("Time"); plt.grid(True); plt.legend(); plt.tight_layout()
    plt.show()
except Exception as e:
    print("Plot skipped:", e)
```
